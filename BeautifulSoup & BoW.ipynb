```python
#TASK1
from time import sleep
from tqdm import *

PAUSE=0.1
URL ='https://pikabu.ru/search?t=%s%%2CТекст&n=2&r=7&page=%d'
STORIES_PER_PAGE=16
TARGET_N=100
DELIMETERS=[":",",","."," ",")","(","!"]
STRICT_LIMIT=False #Cut corpus if size >TARGET_N
stories=[]

for i in tnrange(TARGET_N//STORIES_PER_PAGE+1):
    for x in soup.find_all('div',{'class':'story__content'}):
        stories.append(x.text.strip())
    sleep(PAUSE)
if STRICT_LIMIT:
    stories=stories[:TARGET_N]
print("Total stories:",len(stories))
print("-"*20)
print(stories[:5])
    
```

```python
from math import *

def split(delimiters, string, maxsplit=0):
    import re
    regexPattern = '|'.join(map(re.escape, delimiters))
    return set(filter(lambda x: x!='', re.split(regexPattern, string, maxsplit)))

def text2words(text: str):
    words = []
    _words = split(DELIMETERS,text)
    for word in _words:
        words.append(word.strip())
    return words
tags=[]
for story in stories:
    tags += text2words(story)
tags=set(tags)
list(tags)[:5]
```

```python
def text2bow(text:str):
    words=text2words(text)
    bow=dict()
    for word in words:
        if word in bow:
            bow[word] += 1
        else:
            bow[word] = 1
    return bow
```

*NOTICE:* Corpus BoW is an array of corpus' text's BoWs

```python
bow=[]
for text in stories:
    bow.append(text2bow(text))
```

```python
def bow_dist(bow1: dict, bow2: dict) -> float: # Will use euclidean distance
    s = 0
    visited=[]
    for key in bow1:
        if key in bow2:
            s += (bow1[key]-bow2[key])**2
        else:
            s += (bow1[key])**2
        visited.append(key)
    for key in bow2:
        if key in visited:
             continue
        else:
            if key in bow1:
                s += (bow1[key]-bow2[key])**2
            else:
                s += (bow2[key])**2
    return sqrt(s)
    
```

```python
min_dist = bow_dist(bow[0],bow[1])
min1, min2 = bow[0], bow[1]
for i in range(1,len(bow),1):
    for j in range(1,len(bow),1):
        if i == j:
            continue
        d = bow_dist(bow[i],bow[j])
        if d < min_dist:
            print(i, j)
            min_dist = d
            min1, min2 = bow[i], bow[j]
```
