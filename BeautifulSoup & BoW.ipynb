# Task 0
Describe, why do we use ?t={}%2CТекст&n=2&r=7 in the cell above?
* ? -- HTTP standard divider for passing parmeters to server
* t={} -- parmeter t is begins from `{}`, which means that it will be replaced when string will be frommated 
* %2CТекст -- `%2C` <=> `,` is putted as char code, so t is `{},Текст` where `{}` is replaced by python's formatter
* & -- HTTP arameter list divder
* n=2 -- Parameter `n` is `2`
* r=7 -- Parameter `r` is `7`
